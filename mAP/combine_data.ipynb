```python
import pandas as pd
```

```python
train_df = pd.read_csv('data/train_pred.csv')
test_df = pd.read_csv('data/test_pred.csv')
```

```python
for i in range(train_df.shape[0]):
    train_df.iloc[i,0] = train_df.iloc[i,0]+1000
```

```python
combined_df = train_df.append(test_df,sort=False)
```

```python
combined_df.to_csv('data/pred.csv',index=False)
```
